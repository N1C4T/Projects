```python
import pandas as pd
```

```python
df = pd.read_csv('bina_listings.csv')
df.sample(5)
```

```python
df.shape
```

```python
df1 = df.drop(['Title', 'URL', 'Binanın növü'], axis = 1)
df1.sample(5)
```

```python
df1['Kateqoriya'].unique()
```

```python
residential_categories = ["Yeni tikili", "Köhnə tikili", "Həyət evi/Bağ evi"]
df1 = df1[df1['Kateqoriya'].isin(residential_categories)]
df1.sample(5)
```

```python
df1['Kateqoriya'].unique()
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
cols = ['Çıxarış', 'İpoteka', 'Təmir', 'Torpaq sahəsi']
for i in cols:
    print(df1[i].unique())
```

```python
df2 = df1.copy()
for i in cols:
    df2[i] = df1[i].fillna(0)

df2.sample(5)
```

```python
df2['Çıxarış'] = df2['Çıxarış'].replace({'var': 1, 'yoxdur': 0}).astype('int64')
df2['Təmir'] = df2['Təmir'].replace({'var': 1, 'yoxdur': 0}).astype('int64')
df2['İpoteka'] = df2['İpoteka'].replace('var', 1).astype('int64')

df2.sample(5)
```

```python
# Cleaning Price
df2["Price(AZN)"] = (df2["Price"].str.replace(r"[^\d]", "", regex=True).astype("Int64"))
df2["Torpaq sahəsi(sot)"] = (df2["Torpaq sahəsi"].str.replace(r"[^\d]", "", regex=True).astype("Int64"))

# Cleaning Area
if "Sahə" in df2.columns:
    df2["Area(m2)"] = (df2["Sahə"].str.extract(r"(\d+)").astype("float"))

# Clean Floor info
if "Mərtəbə" in df2.columns:
    floor_split = df2["Mərtəbə"].str.extract(r"(\d+)\s*/\s*(\d+)")
    df2["Floor"] = floor_split[0].astype("Int64")
    df2["TotalFloors"] = floor_split[1].astype("Int64")

df3 = df2.drop(['Mərtəbə', 'Sahə', 'Price', 'Torpaq sahəsi'], axis = 1) 

df3.sample(5)
```

```python
df3.isnull().sum()
```

```python
df3['Otaq sayı'] = df3['Otaq sayı'].fillna(df3['Otaq sayı'].median())

df3['Torpaq sahəsi(sot)'] = df3['Torpaq sahəsi(sot)'].fillna(0)

floors = ['Floor', 'TotalFloors']
df3[floors] = df3[floors].fillna(1)

df3.sample(5)
```

```python
len(df3['Location'].unique())
```

```python
df3['Location'].value_counts()
```

```python
df3['Price(AZN)'].describe()
```

```python
len(df3[df3['Price(AZN)']<10000])
```

```python
df3.shape
```

```python
df4 = df3[df3['Price(AZN)']>10000]
df4.shape
```

```python
len(df4[df4['Area(m2)']>250])
```

```python
df_area_out = df4[df4['Area(m2)']<250]
```

```python
df5 = df_area_out.copy()
df5['price_per_m2'] = (df5['Price(AZN)']/df5['Area(m2)']).round(2)
df5.sample(5)
```

```python
df5['price_per_m2'].describe()
```

```python
import numpy as np

def remove_pps_otliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Location'):
        m = np.mean(subdf.price_per_m2)
        std = np.std(subdf.price_per_m2)
        df_inline = subdf[(subdf.price_per_m2 > (m - std)) & (subdf.price_per_m2 < (m + std))]
        df_out = pd.concat([df_out, df_inline], ignore_index = True)
    return df_out

df6 = remove_pps_otliers(df5)
df6.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.hist(df6.price_per_m2)
plt.show()
```

```python
df7 = df6.drop(['price_per_m2'], axis = 1)
df7.sample(5)
```

```python
df7.Location.unique()
```

```python
df7['Location'] = df7['Location'].str.replace('\xa0', ' ', regex=False)
df7.Location.unique()
```

```python
df8 = pd.get_dummies(df7, columns=['Kateqoriya', 'Location'], drop_first=False).astype(int)
df8.columns = df8.columns.str.replace('Location_', '', regex=False)
df8.columns = df8.columns.str.replace('Kateqoriya_', '', regex=False)
df8.sample(5)
```

```python
df7.Location.value_counts()
```

```python
df9 = df8.drop(['Həyət evi/Bağ evi', '8-ci kilometr q.', 'Köhnə tikili', 'Çıxarış', 'İpoteka', 'Torpaq sahəsi(sot)', 'Floor', 'TotalFloors'], axis = 1)
df9.sample(5)
```

```python
df9.shape
```

```python
y = df9['Price(AZN)']
X = df9.drop('Price(AZN)', axis = 1)
X.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state = 5)
```

```python
from sklearn.linear_model import LinearRegression
lin_clf = LinearRegression()
lin_clf.fit(X_train, y_train)
lin_clf.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 5)
cross_val_score(lin_clf, X, y, cv = cv)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

def best_model(X, y):
    algos = {
        'LinearRegression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },

        'DecisionTreeRegressor': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter': ['best','random']
            }
        },

        'RandomForestRegressor': {
            'model': RandomForestRegressor(),
            'params': {
                'n_estimators': [50, 100]
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=5)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

best_model(X, y)
```

```python
model = DecisionTreeRegressor(criterion = 'friedman_mse', splitter = 'best')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model1 = LinearRegression(fit_intercept = True)
model1.fit(X_train, y_train)
model1.score(X_test, y_test)
```

```python
def predict_price(location, rooms, area, təmir, building):
    loc_index = np.where(X.columns == location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = rooms
    x[1] = təmir
    x[2] = area
    x[3] = building
    if loc_index > 0:
        x[loc_index] = 1
    
    return model.predict([x])[0].round()
```
